```python
#Solutions to Quiz question 17-25

#Importing relevant libraries
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt

#Loading data
df = pd.read_csv('C:\\Users\\PECU\\Downloads\\energydata_complete.csv')
df


#Checking columns and dropping missing values
df.columns

df = df.dropna()

df

#Checking if missing values have been dropped
df.isnull().sum()


#renaming columns
column_new = {'date':'Date','Appliances':'Appliances','lights':'Lights',
              'T1':'Temperature_kitchen','RH_1':'Humidity_kitchen',
             'T2':'Temperature_livingroom','RH_2':'Humidity_livingroom',
              'T3':'Temparature_laundryroom','RH_3':'Humidity_laundryroom',
             'T4':'Temperature_officeroom','RH_4':'Humidity_officeroom',
             'T5':'Temperature_bathroom','RH_5':'Humidity_bathroom',
             'T6':'Temperature_outside','RH_6':'Humidity_outside',
             'T7':'Temperature_ironingroom','RH_7':'Humidity_ironingroom',
             'T8':'Temperature_teenagerroom','RH_8':'Humidity_teenagerroom',
             'T9':'Temperature_parentsroom','RH_9':'Humidity_parentsroom',
             'T_out':'T_out','Press_mm_hg':'Press_mm_hg','RH_out':'Humidity_outside',
             'Windspeed':'Windspeed','Visibility':'Visibility','Tdewpoint':'Tdewpoint',
             'rv1':'Random_variable1','rv2':'Random_variable2'}

df = df.rename(columns = column_new)

df


#Checking new columns
df.columns

#Checking Shape and description
df.shape

df.describe()


#SimpleLinearRegression and plot on Appliance and Random Variable 1

linear_reg_df = df[['Appliances', 'Random_variable1']].sample(15,random_state=2)
sns.regplot(x = 'Appliances', y = 'Random_variable1', data = linear_reg_df)

#Question 17

#Fitting a linear model between the tempearature in the living room and outside the building
linear_reg_df = df[['Temperature_livingroom', 'Temperature_outside']].sample(15,random_state=2)
sns.regplot(x = 'Temperature_livingroom', y = 'Temperature_outside', data = linear_reg_df)


#Question 18
dff = df.drop(['Date', 'Lights'], axis = 1)
dff


#Multiple linear Regression
#Declaring Inputs and Targets
targets = dff['Appliances']
inputs = dff.drop(['Appliances'],axis = 1)   


#Normalizing and scaling inputs
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
Scaler.fit(inputs,targets)


inputs_scaled = Scaler.transform(inputs)

#Train-Test-Split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(inputs_scaled,targets, test_size =0.3, random_state = 42)

#Performing Regression
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train)

#Score
reg.score(X_train,y_train)

#Coefficients
reg.coef_

#Intercept
reg.intercept_


#Summary
reg_summary = pd.DataFrame(inputs.columns.values,columns = ['Features'])
reg_summary ['Weights'] = reg.coef_
reg_summary 


#Obtaining Predictions
prediction_values = reg.predict(X_test)
prediction_values


#Pred  summary
pred_summary = pd.DataFrame()
pred_summary ['Percentages'] = prediction_values
pred_summary

#MeanAbsoluteError
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, prediction_values)
round(mae,3)


#Question 19
from sklearn.metrics import mean_squared_error

mse = (np.sqrt(mean_squared_error(y_test,prediction_values)))
round(mse,3)


#Question 20
#MeanAbsoluteError
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, prediction_values)
round(mae,3)



#Question 21
from sklearn.metrics import mean_squared_error

mse = np.sqrt(mean_squared_error(y_test,prediction_values))
round(mse,3)


#Question 22
Overfit = Yes 

#Question 23
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha = 0.5)
ridge_reg.fit(X_train,y_train)

#Summary
ridge_reg_summary = pd.DataFrame(inputs.columns.values,columns = ['Features'])
ridge_reg_summary ['Weights'] = ridge_reg.coef_
ridge_reg_summary


from sklearn.metrics import mean_squared_error
mse = np.sqrt(mean_squared_error(y_test,prediction_values))
round(mse,3)


#Question 24
from sklearn.linear_model import Lasso

Lasso_reg = Lasso(alpha = 0.5)
Lasso_reg.fit(X_train,y_train)


#Question 25
Lasso_reg_summary = pd.DataFrame(inputs.columns.values,columns = ['Features'])
Lasso_reg_summary ['Weights'] = Lasso_reg.coef_
Lasso_reg_summary


from sklearn.metrics import mean_squared_error
mse = np.sqrt(mean_squared_error(y_test,prediction_values))
round(mse,3)


#Thanks!
```
